# 03 Compare and contrast mutable and immutable objects in  with examples.

```python
# In , objects are classified as mutable or immutable based on whether their state (value) can be changed after creation. Understanding this distinction is crucial for managing data and writing efficient code.

# ### Mutable Objects

# Definition: Mutable objects can be changed after they are created. You can modify their content without creating a new object.

# Examples:
# 1. Lists:
#    my_list = [1, 2, 3]
#    my_list[0] = 10  # List is mutable; the first element is updated
#    print(my_list)   # Output: [10, 2, 3]

# 2. Dictionaries:
#    my_dict = {'a': 1, 'b': 2}
#    my_dict['a'] = 10  # Dictionary is mutable; value associated with key 'a' is updated
#    print(my_dict)     # Output: {'a': 10, 'b': 2}

# 3. Sets:
#    
#    my_set = {1, 2, 3}
#    my_set.add(4)  # Set is mutable; new element 4 is added
#    print(my_set)  # Output: {1, 2, 3, 4}


# ### Immutable Objects

# Definition: Immutable objects cannot be changed once they are created. Any modification results in the creation of a new object.

# Examples:
# 1. Tuples:
#    my_tuple = (1, 2, 3)
#    # my_tuple[0] = 10  # This would raise an error as tuples are immutable
#    new_tuple = (10,) + my_tuple[1:]  # Create a new tuple with modified content
#    print(new_tuple)  # Output: (10, 2, 3)


# 2. Strings:
#    my_string = "hello"
#    # my_string[0] = "H"  # This would raise an error as strings are immutable
#    new_string = "H" + my_string[1:]  # Create a new string with modified content
#    print(new_string)  # Output: "Hello"


# 3. Integers and Floats:
#    my_int = 10
#    # my_int += 5  # This does not change the original object but creates a new integer
#    print(my_int)  # Output: 10 (original object unchanged)


# ### Comparison

# 1. Mutability:
#    - Mutable: Can be modified in place (e.g., lists, dictionaries, sets).
#    - Immutable: Cannot be modified; changes create a new object (e.g., tuples, strings, integers).

# 2. Memory Management:
#    - Mutable: Changes affect the same object reference.
#    - Immutable: Any change results in a new object, which can affect performance due to object creation.

# 3. Use Cases:
#    - Mutable: Suitable for collections where elements may need to be updated or changed (e.g., maintaining a list of items).
#    - Immutable: Useful for fixed values or data that should not change (e.g., using strings for constant text).

# In summary, mutable objects allow in-place modifications, while immutable objects require creating new objects for changes, impacting performance and memory usage in different ways.
```

# 4 Discuss the different types of operators in  and provide examples of how they are used?

```python
# In , the main types of operators include: arithmetic operators (for basic calculations), comparison operators (for comparing values), logical operators (for combining conditions), assignment operators (for assigning values to variables), membership operators (for checking if a value is within a sequence), identity operators (for checking if two objects are the same), and bitwise operators (for manipulating individual bits in binary numbers). 
# Example Operators and Usage: 

# Arithmetic Operators: 
# + (addition): result = 5 + 3 
# - (subtraction): difference = 10 - 2 
# * (multiplication): product = 4 * 2 
# / (division): quotient = 12 / 3 
# % (modulo): remainder = 10 % 3 
#  (exponentiation): power = 2  3 
# // (floor division): floor_result = 11 // 2 

# Comparison Operators: 
# == (equal to): is_equal = 5 == 5 
# != (not equal to): is_not_equal = 2 != 3 
# < (less than): is_less = 1 < 5 
# > (greater than): is_greater = 10 > 5 
# <= (less than or equal to): is_less_or_equal = 5 <= 5 
# >= (greater than or equal to): is_greater_or_equal = 8 >= 7 

# Logical Operators: 
# and (logical AND): result = (5 > 2) and (3 < 8) 
# or (logical OR): result = (True or False) 
# not (logical NOT): result = not (1 == 2) 

# Assignment Operators: 
# = (simple assignment): x = 10 
# += (add and assign): y += 2 (equivalent to y = y + 2) 
# -= (subtract and assign): z -= 1 
# *= (multiply and assign): a *= 3 

# Membership Operators: 
# in (check if value is in sequence): is_in = "a" in "apple" 
# not in (check if value is not in sequence): is_not_in = "b" not in "apple" 
# Identity Operators: 
# is (check if objects are the same): same_object = x is y 
# is not (check if objects are not the same): not_same_object = x is not y 

# Bitwise Operator:
# Bitwise AND (&): Compares each bit of two numbers. The result is 1 if both corresponding bits are 1; otherwise, it is 0.
# Bitwise OR (|): Compares each bit of two numbers. The result is 1 if at least one of the corresponding bits is 1; otherwise, it is 0.
# Bitwise XOR (^): Compares each bit of two numbers. The result is 1 if the corresponding bits are different; otherwise, it is 0.
# Bitwise NOT (~): Inverts the bits of a number. It changes all 0 bits to 1 and all 1 bits to 0. The result is equivalent to -(n+1).
# Left Shift (<<): Shifts the bits of a number to the left by the specified number of positions. Each left shift operation multiplies the number by 2.
# Right Shift (>>): Shifts the bits of a number to the right by the specified number of positions. Each right shift operation divides the number by 2.

# Identity Operators:
# is Operator: Checks if two variables point to the same object in memory. Returns True if both variables refer to the same object, False otherwise.
# is not Operator: Checks if two variables point to different objects in memory. Returns True if both variables refer to different objects, False otherwise.
```

# 06 How do conditional statements work in ? Illustrate with examples?

```python
# Conditional statements in  are used to perform different actions based on different conditions. The primary conditional statements in  are `if`, `elif`, and `else`. These statements help in executing specific blocks of code depending on whether a condition is true or false.

# Here's a detailed look at how conditional statements work in , illustrated with examples:

# ### 1. Basic `if` Statement

# The `if` statement evaluates a condition and executes the block of code that follows it if the condition is true.

# Syntax:

# if condition:
#     # code to execute if condition is true


# Example:

# age = 18
# if age >= 18:
#     print("You are an adult.")

# Output:

# You are an adult.

# In this example, the condition `age >= 18` is true, so the message "You are an adult." is printed.

# ### 2. `if-else` Statement

# The `if-else` statement provides an alternative block of code to execute when the condition is false.

# Syntax:

# if condition:
#     # code to execute if condition is true
# else:
#     # code to execute if condition is false


# Example:

# age = 16
# if age >= 18:
#     print("You are an adult.")
# else:
#     print("You are a minor.")

# Output:

# You are a minor.

# In this example, the condition `age >= 18` is false, so the message "You are a minor." is printed.

# ### 3. `if-elif-else` Statement

# The `if-elif-else` statement allows you to check multiple conditions. It evaluates each condition in order, and executes the corresponding block of code for the first true condition. If none of the conditions are true, the `else` block is executed.

# Syntax:

# if condition1:
#     # code to execute if condition1 is true
# elif condition2:
#     # code to execute if condition2 is true
# else:
#     # code to execute if neither condition1 nor condition2 is true


# Example:

# score = 85

# if score >= 90:
#     print("Grade: A")
# elif score >= 80:
#     print("Grade: B")
# elif score >= 70:
#     print("Grade: C")
# else:
#     print("Grade: D")

# Output:

# Grade: B

# Here, the score is 85, so the condition `score >= 80` is true, and "Grade: B" is printed. The other conditions are not checked once the correct one is found.

# ### 4. Nested `if` Statements

# You can also have `if` statements inside other `if` statements. This is useful for checking more complex conditions.

# Example:

# temperature = 25
# rainy = False

# if temperature > 20:
#     if not rainy:
#         print("It's a warm and dry day. Perfect for a walk!")
#     else:
#         print("It's warm, but it's raining. Consider an umbrella.")
# else:
#     print("It's either cold or rainy. Stay indoors.")

# Output:

# It's a warm and dry day. Perfect for a walk!

# In this example, the inner `if` checks if it's raining only if the temperature is above 20.

# ### 5. Using Logical Operators

# You can combine multiple conditions using logical operators like `and`, `or`, and `not`.

# Example:

# temperature = 22
# humidity = 70

# if temperature > 20 and humidity > 50:
#     print("It's a warm and humid day.")
# else:
#     print("The weather is not warm and humid.")

# Output:

# It's a warm and humid day.

# Here, both conditions (temperature > 20 and humidity > 50) are true, so the message "It's a warm and humid day." is printed.

# ### Summary

# - `if`: Executes a block of code if a condition is true.
# - `if-else`: Provides an alternative block of code to execute if the condition is false.
# - `if-elif-else`: Checks multiple conditions and executes the block of code corresponding to the first true condition.
# - Nested `if` statements: Allows more complex condition checking.
# - Logical operators: Combine multiple conditions to form more complex expressions.

# Conditional statements are fundamental in controlling the flow of your program based on varying conditions.
```
